```python
from dataset import RawDataset
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
unit = "VG5"
rds = RawDataset(unit=unit, load_synthetic=False, load_training=True)
data_train = rds.data_dict["train"].measurements

scaler = StandardScaler()
pca = PCA(n_components=None)
pca.fit(scaler.fit_transform(data_train))

# plt.tight_layout()

fig, ax = plt.subplots(1, 1, figsize=(22, 18))
sns.heatmap(data_train.corr(), cmap="vlag", xticklabels=True, yticklabels=True)
ax.set_title(f"Correlation map for {unit}")

components = pd.DataFrame(pca.components_, columns=data_train.columns)
fig, ax = plt.subplots(1, 1, figsize=(18, 14))
sns.heatmap(components.transpose(), cmap="vlag", xticklabels="auto", yticklabels=True)
ax.set_title(f"Principal components (eigenspace) map for {unit}")
```
